```python
def main():
    # 定义基坑计算必须的参数
    E = 60E9
    h = 1
    I = h ** 3 / 12
    D = 1
    v = 0.3

    print(E * I)

    L1 = L2 =16.4
    H1 = H2 = 12.4
    # L2 = 24
    
    B = 46.9

    supportCount = 1

    # L1, H1, L2, H2 = symbols('L_1 H_1 L_2 H_2')
    Ns = [symbols(f'N_{i + 1}') for i in range(supportCount)]
    Nsn = {i: 0 for i in Ns}
    ds = [i for i in [0]]
    z, s = symbols('z s', positive=True)

    # 定义地连墙
    concreteMaterial = ConcreteMaterial('钢筋混凝土', 25, E, v)
    leftWall = UndergroundDiaphragmWall(L1, h, concreteMaterial)
    # rightWall = UndergroundDiaphragmWall(L2, h, concreteMaterial)

    # 定义土层与钻孔
    intervals = []
    soilMaterial4 = SoilMaterial('Mudstone ', 25.9E3, 10E7, phi=30, c=300)
    intervals.append({
        'top': 0,
        'bottom': 16.4,
    })

    boreHole = BoreHole([soilMaterial4], intervals)

    # 定义水平支撑
    # steelSupportMaterial = SupportMaterial('钢支撑', 25E3, 35E9, 1.2)
    concreteSupportMaterial = SupportMaterial('混凝土支撑', 25E3, 10E8, 1)

    supports = [HorizontalSupport(concreteSupportMaterial, B)]

    NEA = [i.Material.EA for i in supports]

    

    NN = {
        'Ns': Ns,
        'NEA': NEA  
    }
    # solver = EnergyMethodSolver(z, foundationPit, 12)
    # result, wl, wr = solver.Solve()

    for i in range(7):
        L2 = 16.4 - i / 2
        rightWall = UndergroundDiaphragmWall(L2, h, concreteMaterial)
        foundationPit = FoundationPit(leftWall, rightWall, H1, H2, supports,
                                        supportCount, ds, B, D, boreHole, 
                                        leftOverLoad=10E3, rightOverLoad=10E3,
                                        leftStrengthLoad=30E3, rightStrengthLoad=30E3)
        solver = EnergyMethodSolver(z, foundationPit, 12)
        result, wl, wr = solver.Solve()
        return result, wl, wr



```

```python
result, wl, wr = main()
```

```python
obj = {
    'result': result,
    'wl': dict(wl.as_coefficients_dict()),
    'wr': dict(wr.as_coefficients_dict()),
    'symbol': 'z'
}

dic = {}
for key in result:
    dic[str(key)] = result[key]

dic

# import json
# json.dumps(obj)
```
